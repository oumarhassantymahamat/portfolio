# PROJET 3 : équation de poisson en dimension 2

On étudie dans ce projet la mise en oeuvre de schémas numériques pour l'équation de poisson en dimension 2.
On s'interresse d'abort à l'équation suivante :

\begin{cases}
-\Delta u(x, y) = f(x, y) & \text{pour tout } (x, y) \in ]0, 1[ \times ]0, 1[ \\
u(x, y) = 0 & \text{si } x \in \{0, 1\} \text{ ou } y \in \{0, 1\}
\end{cases}

Avec $\Delta u(x, y) = \partial^2 u(x, y) + \partial^2 u(x, y)$ est appelé le Laplacien de $u$. $h = \frac{1}{N+1}$ et on souhaite approcher la solution $u$ de l’équation aux points $(x_{i},y_{j}) = (ih,jh)$ par $u_{i,j}$ pour $N+1$ $i=1..N$ et $j=1..N$.

# PARTIE I

### 1- Système linéaire obténu en discrétisant l'équation par différences finies

Une méthode d'approximation efficace pour la dérivée seconde par rapport à x, de l'équation de poisson à une dimension notée u''(x), est donnée par l'expression $\frac{{u_{i+1} - 2u_i + u_{i-1}}}{{h^2}}$, avec $h = \frac{1}{{N+1}} $. En étendant cette approche à la dérivée seconde par rapport x et y en même temps dans l'expression $ \Delta u(x, y) = \frac{{\partial^2 u(x, y)}}{{\partial x^2}} + \frac{{\partial^2 u(x, y)}}{{\partial y^2}} $ de l'équation de poisson à 2 dimension, nous obtenons une forme améliorée : $ \frac{{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}}{{h^2}} + \frac{{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}}{{h^2}} $, où les indices i  et  j  correspondent respectivement aux pas de discrétisation selon les axes x  et  y, plus précisement i = 1...N et j = 1...N

Le shéma aux différences finis de notre équation de poisson à deux dimension nous donne :
\begin{cases}
-(\frac{{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}}{{h^2}} + \frac{{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}}{{h^2}}) = f_{i,j}  \\
u(0, y) = u(1, y) = u(x, 0) = u(x, 1) = 0 
\end{cases}

Exprimons le schéma au différences finies sous la forme $AU = b$ où $A$ est une matrice carrée  $N^2 \times N^2$.

### 2 - Existence et Unicité de la solution solution du système linéaire obtenu

Pour montrer l'existence et l'unicité de la solution du système linéaire obtenu à partir du schéma aux différences finies de notre équation, nous allons utiliser le théorème de Lax-Milgram.

Dans notre démarche, notre objectif est de garantir que notre problème satisfait toutes les conditions du théorème de Lax-Milgram. Cela nous permettra d'affirmer avec certitude l'existence et l'unicité de la solution.

### Théorème de Lax-Milgram

Soit $H$ un espace de Hilbert, $a : H \times H \rightarrow \mathbb{R}$ une forme bilinéaire continue et coercive et $L$ une forme linéaire continue sur $H$. Alors il existe un unique $u \in H$ solution de :  $a(u,v) = L(v)$
pour tout $v \in H$.

Appliquons le théorème de Lax-Milgram précédemment énoncé pour démontrer l'existence et l'unicité de la solution.

Le théorème de Lax-Milgram est un résultat fondamental en analyse fonctionnelle qui va nous permettre d'établir l'existence et l'unicité de solutions pour notre problème.


Pour appliquer le théorème de Lax-Milgram à notre équation, on doit d'abord définir un espace fonctionnel approprié sur lequel se trouve la solution. En général, c'est un espace de Sobolev noté $H^1(\Omega)$, où $( \Omega )$ est le domaine de définition de l'équation.

l'espace de Sobolev $( H^1(\Omega) )$ ici dans notre cas est défini comme suit : $ H^1(\Omega)$ = $\{ u \in L^2(\Omega) : \nabla u \in L^2(\Omega) \} $
où $( L^2(\Omega) ) $ est l'espace des fonctions carré-intégrables sur le domaine $( \Omega )$ et $( \nabla u )$ est le gradient de ( u ).


Ensuite, on définit deux formes bilinéaires $( a(u, v) )$ et $( L(v) )$ qui caractérisent l'équation, de sorte que l'équation puisse être écrite sous la forme :
$a(u, v) = L(v)$

Pour l'équation de Poisson la forme bilinéaire $( a(u, v) )$ est définie comme suit :
$ a(u, v) = \int_{\Omega} \nabla u \cdot \nabla v \, dx \, dy $ où $( u )$ et $( v )$ sont des fonctions test appartenant à l'espace $( H^1(\Omega) )$, et $( \Omega )$ est le domaine où l'équation de Poisson est définie.
    
Cette forme bilinéaire mesure la similarité entre les gradients de deux fonctions $( u )$ et $( v )$ sur le domaine $( \Omega )$. Elle est dérivée de l'équation de Poisson elle-même en appliquant une intégration par parties.

La forme linéaire $( L(v) )$ est définie comme suit :$ L(v) = \int_{\Omega} f v \, dx \, dy $ où $( f(x, y) )$ est une fonction donnée (le terme source de l'équation de Poisson) et $( v )$ est une fonction test appartenant à $( H^1(\Omega) )$.
    
Cette forme linéaire représente l'effet du terme source $( f )$ sur les fonctions test $( v )$ dans le domaine $( \Omega )$.

En utilisant ces formes bilinéaire et linéaire, l'équation de Poisson peut être formulée de manière variationnelle comme suit :
$ \text{Trouver } u \in H^1(\Omega) \text{ tel que } a(u, v) = L(v) \text{ pour tout } v \in H^1(\Omega) $


Maintenant, en vérifiant les conditions du théorème de Lax-Milgram, on peut montrer que la solution $( u )$ existe et est unique dans $( H^1(\Omega) )$.

Les conditions essentielles pour appliquer le théorème sont la coercivité de la forme bilinéaire $( a(u, v) )$ et sa continuité, ainsi que la linéarité et la continuité de la forme linéaire $( L(v) )$. Une fois ces conditions vérifiées, le théorème garantit l'existence et l'unicité de la solution $( u )$.

Pour montrer que la forme bilinéaire $(a(u, v))$ associée à l'équation de Poisson est coercive, nous devons démontrer qu'il existe une constante $( \alpha > 0 )$ telle que :

$a(u, u) \geq \alpha | u |_{H^1(\Omega)}^2 $

pour toute fonction $( u )$ appartenant à l'espace de Sobolev $( H^1(\Omega) )$.

La forme bilinéaire $(a(u, v))$ est définie comme :

$ a(u, v) = \int_{\Omega} \nabla u \cdot \nabla v \, dx \, dy $

Pour montrer la coercivité, nous utilisons l'inégalité de Poincaré, qui stipule que pour toute fonction \( u \) dans $( H^1(\Omega) )$, il existe une constante $( C )$ dépendant uniquement de $( \Omega )$ telle que :

$ \| u \|_{L^2(\Omega)} \leq C \| \nabla u \|_{L^2(\Omega)} $

Nous pouvons réécrire la forme bilinéaire $(a(u, u))$ comme :

$ a(u, u) = \| \nabla u \|_{L^2(\Omega)}^2 $

Ainsi, en utilisant l'inégalité de Poincaré, nous avons :

$ a(u, u) = \| \nabla u \|_{L^2(\Omega)}^2 \geq \frac{1}{C} \| u \|_{L^2(\Omega)}^2 $

alors la forme bilinéaire $(a(u, u))$ est coercive avec $( \alpha = \frac{1}{C} )$, ce qui prouve que la forme bilinéaire associée à l'équation de Poisson est coercive.

La forme bilinéaire $( a(u, v) )$ associée à l'équation de Poisson est continue, s'il existe une constante $( M > 0 )$ telle que :

$ |a(u, v)| \leq M \| u \|_{H^1(\Omega)} \| v \|_{H^1(\Omega)} $

pour toute paire de fonctions $( u, v )$ appartenant à l'espace de Sobolev $( H^1(\Omega) )$.

La forme bilinéaire $( a(u, v) )$ est définie comme :

$ a(u, v) = \int_{\Omega} \nabla u \cdot \nabla v \, dx \, dy $

par l'inégalité de Cauchy-Schwarz on a :

$ |a(u, v)| = \left| \int_{\Omega} \nabla u \cdot \nabla v \, dx \, dy \right| \leq \left( \int_{\Omega} |\nabla u|^2 \, dx \, dy \right)^{1/2} \left( \int_{\Omega} |\nabla v|^2 \, dx \, dy \right)^{1/2} $


On sait que $( \nabla u )$ et $( \nabla v )$ sont tous deux dans $( L^2(\Omega) )$, nous pouvons utiliser la norme $( L^2 )$ pour majorer les termes ci-dessus.

$ |a(u, v)| \leq \| \nabla u \|_{L^2(\Omega)} \| \nabla v \|_{L^2(\Omega)} $

Nous savons également que les normes \( L^2 \) sont équivalentes aux normes \( H^1 \), donc :

$ \| \nabla u \|_{L^2(\Omega)} \leq C_1 \| u \|_{H^1(\Omega)} $
$ \| \nabla v \|_{L^2(\Omega)} \leq C_2 \| v \|_{H^1(\Omega)} $

Où $( C_1 )$ et $( C_2 )$ sont des constantes positives dépendant uniquement de $( \Omega )$. En remplaçant dans notre expression précédente, nous obtenons :

$ |a(u, v)| \leq C_1 C_2 \| u \|_{H^1(\Omega)} \| v \|_{H^1(\Omega)} = M \| u \|_{H^1(\Omega)} \| v \|_{H^1(\Omega)} $

Donc, la forme bilinéaire $( a(u, v) )$ est continue avec $( M = C_1 C_2 )$.

La forme linéaire $( L(v) )$ associée à l'équation de Poisson est continue, s'il existe une constante $( M' > 0 )$ telle que :

$|L(v)| \leq M' \| v \|_{H^1(\Omega)} $

pour toute fonction $( v )$ appartenant à l'espace de Sobolev $( H^1(\Omega) )$.

La forme linéaire $( L(v) )$ est définie comme :

$ L(v) = \int_{\Omega} f v \, dx \, dy $

où $( f(x, y) )$ est une fonction donnée .

Par inégalité de Cauchy-Schwarz :

$ |L(v)| = \left| \int_{\Omega} f v \, dx \, dy \right| \leq \left( \int_{\Omega} |f|^2 \, dx \, dy \right)^{1/2} \left( \int_{\Omega} |v|^2 \, dx \, dy \right)^{1/2} $

Puisque $( f )$ et $( v )$ sont toutes deux dans $( L^2(\Omega) )$, gràce à la norme $( L^2 )$ on peux majorer les termes ci-dessus. 

$ |L(v)| \leq \| f \|_{L^2(\Omega)} \| v \|_{L^2(\Omega)} $

Nous savons que les normes $( L^2 )$ sont équivalentes aux normes $( H^1 )$, donc :
$ \| f \|_{L^2(\Omega)} \leq C_3 \| f \|_{H^1(\Omega)} $

Où $( C_3 )$ est une constante positive dépendant uniquement de $( \Omega )$. En remplaçant dans notre expression précédente, on a :
$ |L(v)| \leq C_3 \| f \|_{H^1(\Omega)} \| v \|_{H^1(\Omega)} $

Ainsi, la forme linéaire $( L(v) )$ est continue avec $( M' = C_3 \| f \|_{H^1(\Omega)} )$, ce qui conclut la démonstration. 

Donc par le théorème de Lax-Milgram la solution u existe et est unique.

### 3 - Programme de la resolution du système

On utilise les matrices A et le vecteur b obtenus grâce au schéma aux différences finies précédemment défini. Pour tester la méthode, on applique ces matrices et ce vecteur à la fonction $f(x, y) = -2\sin(\pi x)\sin(\pi y)$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# definition de la fonction du test
def f_t(x, y):
    return -2*np.sin(np.pi*x)*np.sin(np.pi*y)
```

```python
L = 1 #longueur du domaine
N = 50 #nombre de points de discrétisation
h = 1/(N + 1)  #taille du pas
x = np.linspace(h,L-h,N) #points de discrétisation
y = x
X, Y = np.meshgrid(x,y)
Z_table = f_t(X, Y)
# affichage de notre fonction de test en 2D
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z_table,rstride=1,cstride=1,cmap='viridis',edgecolor='none')
ax.view_init(40,35) # rotation de 35 degrés autour de l'axe 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
# on prend n petit pour visualiser la matrice de notre système
n = 10
# Montage et visualisation de la matrice A
diag0 = np.diag(np.ones(n,),k=0)
diag1 = np.diag(np.ones(n-1,),k=1)
diag2 = np.diag(np.ones(n-1,),k = -1 )
A = (1/h**2)*(4*diag0 - diag1 - diag2)
M = np.kron(np.eye(n), A)
M = M - np.diag(np.ones((n)**2-(n-1), ), n-1)
M = M - np.diag(np.ones((n)**2-(n-1), ), -(n-1))
plt.spy(M)
```

```python
Z = f_t(X, Y).flatten()

# création du vecteur b de notre système
b = [Z[i] for i in range(0, N**2)]  #vecteur b

# Montage et visualisation de la matrice A
diag0 = np.diag(np.ones(N,),k=0)
diag1 = np.diag(np.ones(N-1,),k=1)
diag2 = np.diag(np.ones(N-1,),k = -1 )
A = (1/h**2)*(4*diag0 - diag1 - diag2)
M = np.kron(np.eye(N), A)
M = M - np.diag(np.ones((N)**2-(N-1), ), N-1)
M = M - np.diag(np.ones((N)**2-(N-1), ), -(N-1))
#résolution du système et visualisation du resultat
U = np.linalg.solve(M,b)
U_matrix = np.reshape(U, (N, N))
# figure
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,U_matrix,rstride=1,cstride=1,cmap='viridis',edgecolor='none')
ax.view_init(40,35) # rotation de 35 degrés autour de l'axe
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

# PARTIE II

On s'interesse dans cette partie à l'équation suivante :

\begin{cases}
-\Delta u(x, y) + c\frac{{\partial u(x, y)}}{{\partial x2}} = f(x, y) &  (x, y) \in ]0, 1[^2 \\
u(x, y) = 0 & \text{si } x \in \{0, 1\} \text{ ou } y \in \{0, 1\}
\end{cases}

On note $u$ la solution de notre équation, et on suppose que $f \in C([0,1]^2)$ et $u \in C^4([0,1])$ . On admettra que si $f ≤ 0$, alors $u ≤ 0$.
Dans cette partie on veux étudier deux schémas numériques pour approcher $u(x_{i},y_{j})$ par $u_{i,j}$. On notera $f_{i,j} = f(x_{i},y_{j})$. 

Le premier schéma, appelé schéma 1 est le suivant :
$  \frac{4}{{h^2}}u_{i,j} - \left(\frac{1}{{h^2}} + \frac{c}{{2h}}\right)u_{i-1,j} - \left(\frac{1}{{h^2}} - \frac{c}{{2h}}\right)u_{i+1,j} - \frac{1}{{h^2}}u_{i,j-1} - \frac{1}{{h^2}}u_{i,j+1} = f_{i,j}$

Le sécond  schéma, appelé schéma 2 est le suivant : $ (\frac{{4}}{{h^2}} + \frac{{c}}{{h}}) u_{i,j} - (\frac{{1}}{{h^2}} + \frac{{c}}{{h}}) u_{i-1,j} - \frac{{1}}{{h^2}} u_{i+1,j} - \frac{{1}}{{h^2}} u_{i,j+1} - \frac{{1}}{{h^2}} u_{i,j-1}= f_{ij} $

### 1- Codons ces deux schémas

On commence d'abord par exprimer les schéma au différences finies sous la forme $AU = b$ où $A$ est une matrice carrée  $N^2 \times N^2$ dans les deux cas. 
Nous allons directement ecrire le code de ces deux schémas

```python
# definition de la fonction
def f(x,y):
    return x*(x-1)*y*(1-y)
```

```python
# schéma 1
def schema_1(N, c, L = 1):
    h = 1/(N + 1)  #taille du pas
    x = np.linspace(h,L-h,N) #points de discrétisation
    y = x
    X, Y = np.meshgrid(x,y)
    Z = f(X, Y).flatten()
    # création du vecteur b de notre système
    b = [Z[i] for i in range(0, N**2)]  #vecteur b

    # Montage et visualisation de la matrice A
    diag0 = np.diag(np.ones(N,),k=0)
    diag1 = np.diag(np.ones(N-1,),k=1)
    diag2 = np.diag(np.ones(N-1,),k = -1 )
    A = (1/h**2)*(4*diag0 - (1+((c*h)/2))*diag2 - (1+((c*h)/2))*diag1)
    M = np.kron(np.eye(N), A)
    M = M - np.diag(np.ones((N)**2-(N-1), ), N-1)
    M = M - np.diag(np.ones((N)**2-(N-1), ), -(N-1))
    #résolution du système et visualisation du resultat
    U = np.linalg.lstsq(M,b,rcond=None)[0]
    return U
```

```python
# Schéma 2
def schema_2(N, c, L = 1):
    h = 1/(N + 1)  #taille du pas
    x = np.linspace(h,L-h,N) #points de discrétisation
    y = x
    X, Y = np.meshgrid(x,y)
    Z_table = f(X, Y)
    Z = f(X, Y).flatten()
    # création du vecteur b de notre système
    b = [Z[i] for i in range(0, N**2)]  #vecteur b

    # Montage et visualisation de la matrice A
    diag0 = np.diag(np.ones(N,),k=0)
    diag1 = np.diag(np.ones(N-1,),k=1)
    diag2 = np.diag(np.ones(N-1,),k = -1 )
    A = (1/h**2)*((4+c*h)*diag0 - (1+c*h)*diag2 - diag1)
    M = np.kron(np.eye(N), A)
    M = M - np.diag(np.ones((N)**2-(N-1), ), N-1)
    M = M - np.diag(np.ones((N)**2-(N-1), ), -(N-1))
    #résolution du système et visualisation du resultat
    U = np.linalg.solve(M,b)
    return U
```

$( h_1, h_2 )$ sont les valeurs de $( h )$ correspondantes.

- Pour $( h_2 = \frac{1}{10} )$, $( e_2 = 0.061390992697849756 )$
